## Expenses

```python
total = table["Money Owed"].sum()
food = 80+58
alc = 97+70.48+57+49
pizza = 65.44
```

```python
# Initialising the array
foodp = []
alcp = []
pizzap = []
```

```python
for index, row in table.iterrows():
    if row["Food"] == 'Y':
        foodp.append(row["Name"])    
```

```python
for index, row in table.iterrows():
    if row["Alc"] == 'Y':
        alcp.append(row["Name"])
```

```python
for index, row in table.iterrows():
    if row["Pizza"] == 'Y':
        pizzap.append(row["Name"])
```

```python
foodp = pd.DataFrame(foodp, columns=['Name'])
foodpp = food/len(foodp.index)
foodp["Owes"] = foodpp
```

```python
alcp = pd.DataFrame(alcp, columns=['Name'])
alcpp = alc/len(alcp.index)
alcp["Owes"] = alcpp
```

```python
pizzap = pd.DataFrame(pizzap, columns=['Name'])
pizzapp = pizza/len(pizzap.index)
pizzap["Owes"] = pizzapp
```

```python
totalintdf = foodp.set_index('Name').add(alcp.set_index('Name'), fill_value=0).reset_index()
totalintdf
```

```python
totaldf = totalintdf.set_index('Name').add(pizzap.set_index('Name'), fill_value=0).reset_index()
df1 = table[['Name', 'Money Owed']]
totaldf = totaldf.set_index('Name').add(df1.set_index('Name'), fill_value=0).reset_index()
totaldf
```

```python
totaldf["final price"] =round( totaldf["Money Owed"] - totaldf["Owes"], 2)
totaldf
```
